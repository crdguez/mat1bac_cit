```python
from sympy import init_session
from sympy import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Modifica datos

cota_inf_x = -10.
cota_sup_x = 10.
num_puntos = 20
x = np.linspace(cota_inf_x,cota_sup_x,num_puntos)

# Modificar la funci√≥n

# y = x**2
y = x**(-1)

plt.plot(x,y,'ro')
plt.grid(True)
plt.axis('equal')
plt.show()

pd.DataFrame({'x':x,'y':y})



```
