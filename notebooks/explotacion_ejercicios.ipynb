```python
import pandas as pd
import numpy as np
from docparts import escribir_ejercicios, escribir_fin, escribir_preambulo, generar_tex
from IPython.display import Markdown as md
from IPython.display import display
```

```python
fichero = 'ejercicios'
n_ficheros = 4

n_ejercicios = 15

df = pd.concat([pd.read_csv(fichero + str(n + 1)+ '.csv', index_col=False) for n in range(n_ficheros)],sort=True)

df = df.reset_index()
```

```python
df
```

```python
df_ejercicios = df.loc[np.random.choice(df.index, n_ejercicios, replace=False)].sort_values('n_ejercicio')
```

```python
df_ejercicios
```

```python
fichero = 'autoevaluacion'
titulo = 'Autoevaluación'
tipo = 'ejerc'
n_pdf = ['C']

def exportar_pdf(fichero, titulo, tipo, n_pdf) : 
    titulo = titulo + n_pdf[0]
    if tipo != 'ejercicios' : df_ejercicios.n_columnas = 1 

    escribir_preambulo(fichero, titulo, tipo)
    for s in df_ejercicios.groupby('n_ejercicio').count().index : 
        display(md("**Ejercicio: **" + s ))
        display(df_ejercicios[df_ejercicios.n_ejercicio == s])
        escribir_ejercicios(df_ejercicios[df_ejercicios.n_ejercicio == s],fichero)

    escribir_fin(fichero)
    
exportar_pdf(fichero, titulo, tipo, n_pdf)
```
