```python
# Generador de prueb: Cambia los parámetros 

fichero = 'parc2bis'
titulo = 'Parcial 2ª evaluación'
tipo = 'ejerc'  # Todo lo que sea diferente a 'ejercicios' lo generará

n_pdf = list(string.ascii_uppercase)[:4] # Pruebas diferentes a generar

if tipo != 'ejercicios' :
 
    #Construcción de lista de ejercicios
    numero = 2 # Número de ejercicios por defecto de cada ejercicio
    
    lista1 = np.unique(df.n_ejercicio.tolist())
    lista2 =[]
    for i in range(lista1.size) : lista2.append(numero)
    lista = list(zip(lista1,lista2))
    
    # Si queremos poner nuestra propia lista: lista =[('texto_nejercicio',nº de ejercicios)], si no comentar la linea siguiente
    # lista = [('ex01e01', 1),('ex01e02', 1),('ex01e03', 1),('ex01e05', 2),('ex01e06', 1),('ex01e08', 1)]
    lista = [('ex11e01', 1),('ex11e04b', 4),('ex11e05', 1),('ex11e06b', 1),('ex11e07', 1)]
    
    for s in n_pdf:
        random.shuffle(lista)
        df_ejercicios = pd.concat(df.loc[np.random.choice(df[df.n_ejercicio==s[0]].index, s[1], replace=False)] for s in lista)
        exportar_pdf(df_ejercicios,fichero, titulo, tipo, letra = s, soluciones = False)
        exportar_pdf(df_ejercicios,fichero, titulo, tipo, letra = s, soluciones = True)

else :
    exportar_pdf(df, fichero, titulo, tipo)
```
